```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('Advertising.csv')
```

```python
# Iniciando a Análise Exploratória de Dados (EDA)
```

```python
base.head()
```

```python
base.info()
```

```python
base.describe()
```

```python
# O conjunto de dados apresenta variáveis numéricas contínuas, sem valores ausentes, o que facilita a etapa de modelagem.
```

```python
# Verificando numerosidade de valores nulos
base.isnull().sum()
```

```python
# Removendo coluna desnecessária
base = base.drop(columns=['Unnamed: 0'])
```

```python
# Distribuição das variáveis
base.hist(bins=20, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
# Observa-se que algumas variáveis apresentam distribuição assimétrica, o que pode influenciar o desempenho do modelo.
```

```python
# Fazendo Boxplot para identificar outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=base)
plt.title('Boxplot das Variáveis')
plt.show()
```

```python
# TV e Newspaper apresentam maior variabilidade. 
```

```python
# Relação entre as variáveis independentes e Sales
sns.pairplot(base)
plt.show()
```

```python
# A análise visual sugere uma relação linear positiva entre investimento em TV e vendas, enquanto outras variáveis apresentam impacto menos evidente.
```

```python
# Correlação 
plt.figure(figsize=(8, 6))
sns.heatmap(base.corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# A matriz de correlação mostra que a variável TV possui maior correlação com vendas, indicando maior poder explicativo no modelo de regressão múltipla.
```

```python
# A análise exploratória permitiu compreender o comportamento das variáveis, identificar relações relevantes e fornecer subsídios para a construção do modelo de regressão múltipla, reduzindo decisões arbitrárias durante a modelagem.
```
